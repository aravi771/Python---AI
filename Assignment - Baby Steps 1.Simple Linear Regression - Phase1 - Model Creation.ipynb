```python
# pandas is a library & it contains functions related to the database

import pandas as pan
```

```python
# read the database file and assign it to dataset

data=pan.read_csv("Salary_Data.csv")
```

```python
# Data Preprocessing

data
```

```python
# Input Split

YE=data[["YearsExperience"]]
YE
```

```python
# Output Split

Sal=data[["Salary"]]
Sal
```

```python
# sklearn is the library that support ML & split the training  set & test set

from sklearn.model_selection import train_test_split
```

```python
#call the function that was imported by passsing parameter for input variable, output variable, test size, random state & assign it to a variable for test & train

X_tr,X_tt,Y_tr,Y_tt=train_test_split(YE,Sal,test_size=0.2,random_state=0)
```

```python
# import the class linear regression

from sklearn.linear_model import LinearRegression

# assign the class to a variable

regressor=LinearRegression()

# call the fit function which is used for substitution

regressor.fit(X_tr,Y_tr)
```

```python
# checking the weight by using coef (used to calculate the weight)

weight=regressor.coef_
weight
```

```python
# checking the bais by using intercept (used to calculate the bais)

bais=regressor.intercept_
bais
```

```python
# evaluate the model by using predict function

Y_predi=regressor.predict(X_tt)
```

```python
# find r2 score

from sklearn.metrics import r2_score
r_sco=r2_score(Y_tt,Y_predi)
```

```python
#check the r2 score
r_sco
```

```python
# naming & saving the finalized model

import pickle
filename="FML.sav"
```

```python
# assigning the model into the saved file

pickle.dump(regressor,open(filename,'wb'))
```

```python
# checking the saved model

load_model=pickle.load(open("FML.sav",'rb'))
exp=int(input("Enter Your Years Of Experience:"))
final=load_model.predict([[exp]])
final
```
